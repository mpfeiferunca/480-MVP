```python
import pandas as pd

def readin():
    csv = pd.read_csv('SpecializedReachsml.csv')
    count = 0
    while(count < len(csv)):
        
        count = count + 1
    #print(count)
    csv = csv.loc[:, ~csv.columns.str.contains('^Unnamed')]
    print(csv)
   

    return csv
    
CSV = readin()
```
